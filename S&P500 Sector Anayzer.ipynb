```python
def datacollecter(ticker):
    data = tv_access.get_hist(ticker,"SP",n_bars = 252).close
    return data
```

```python
SP_sectors = ["S5COND","S5CONS","S5HLTH","S5INDU","S5INFT","S5MATR","S5REAS","S5TELS","S5UTIL","SPF","SPN"]
```

```python
all_sectors = pd.DataFrame()
for sector in SP_sectors:
    all_sectors[sector] = datacollecter(sector)
all_sectors.head(10)
```

```python
sector_log_returns = np.log(all_sectors/all_sectors.shift(1))
```

```python
mean_returns = sector_log_returns.mean()*days
mean_returns.columns = ["Mean Returns"]
st_devs = np.sqrt(sector_log_returns.var()*days)
st_devs.columns = ["Volatility"]
sharpe = (mean_returns-risk_free)/st_devs
```

```python
summary_table = pd.DataFrame({'Return': mean_returns, 'Volatility': st_devs,"Sharpe Ratio" : sharpe})
summary_table
```
